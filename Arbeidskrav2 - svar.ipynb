```python
# Oppg 5

import numpy as np

def figur(a, b):
    """Beregning av areal og ytre omkrets for en figur kombinert av en trekant og en halvsirkel.
    a og b (float) er innverdi.
    areal_tot og omkrets_tot (float) er utverdi.
    """
    r = a / 2

    # Arealer
    areal_halv = np.pi * r**2 / 2          # halvsirkel
    areal_tre = b * a / 2                  # trekant
    areal_tot = areal_halv + areal_tre
    
    # Ytre  omkrets
    omkrets_halv = np.pi * r               # halvsirkel
    hypotenus = np.sqrt(a**2 + b**2)  
    omkrets_tot = omkrets_halv + b + hypotenus
    
    return areal_tot, omkrets_tot

# brukerinput
a = float(input('Skriv inn a: '))
b = float(input('Skriv inn b: '))

# kaller på funksjonen "figur"
areal, omkrets = figur (a,b)

print(f'Areal: {areal:.2f}')
print(f'Ytre omkrets: {omkrets:.2f}')
```
